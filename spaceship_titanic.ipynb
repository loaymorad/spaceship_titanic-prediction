## Feature Engineering

```python
df.head()
```

```python
# adding new feature
df['HomeDestination'] = df['HomePlanet'] + '_' + df['Destination']

df = pd.get_dummies(df, columns=['HomeDestination'], drop_first=True)
```

## Data Cleaning

```python
df = df.drop(['Name', 'HomePlanet', 'Destination'], axis='columns')
df.head()
```

```python
df.isnull().sum()
```

```python
df['CryoSleep'].fillna(False, inplace=True)
```

```python
df['VIP'].fillna(True, inplace=True)
```

```python
from sklearn.impute import KNNImputer

# Select the columns for KNN imputation
knn_columns = ['Age', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']

# Extract the relevant columns for KNN imputation
df_knn = df[knn_columns]

# Initialize the KNNImputer
imputer = KNNImputer(n_neighbors=5, weights='uniform')

# Apply the KNN imputer
df_knn_imputed = pd.DataFrame(imputer.fit_transform(df_knn), columns=knn_columns)

# Replace the original columns in the dataframe with the imputed data
df[knn_columns] = df_knn_imputed
```

```python
df['Cabin'].fillna(df['Cabin'].mode()[0], inplace=True)
```

## Build Model

```python
from sklearn.model_selection import GridSearchCV

from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.preprocessing import StandardScaler

X = df.drop('Transported', axis=1)
y = df['Transported']

X_scaled = StandardScaler().fit_transform(X)
```

```python
X.head()
```

```python
def find_best_model_using_gridsearchcv(X_scaled, y):
    model_params = {
        'random_forest': {
            'model': RandomForestClassifier(),
            'params' : {
                'n_estimators': [1, 5, 10]
            }
        },
        'logistic_regression' : {
            'model': LogisticRegression(solver='liblinear', multi_class='auto'),
            'params': {
                'C': [1, 5, 10]
            }
        },
        'decision_tree': {
            'model': DecisionTreeClassifier(),
            'params': {
                'criterion': ['gini', 'entropy']
            }
        }
    }
    scores = []

    for model_name, mp in model_params.items():
        clf = GridSearchCV(mp['model'], mp['params'], cv=5, return_train_score=False)
        clf.fit(X, y)
        scores.append({
            'model': model_name,
            'best_score': clf.best_score_,
            'best_params': clf.best_params_
        })

    df = pd.DataFrame(scores, columns=['model','best_score','best_params'])
    return df

find_best_model_using_gridsearchcv(X, y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# use logisic regresssion
model = LogisticRegression(solver='liblinear', multi_class='auto')
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression

param_grid = {
    'model__C': [0.01, 0.1, 1, 10, 100],
    'model__l1_ratio': [0.1, 0.5, 0.7, 0.9, 1]
}

scaler = StandardScaler()
model = LogisticRegression(penalty='elasticnet', solver='saga', max_iter=500)

pipeline = Pipeline(steps=[('scaler', scaler), ('model', model)])

grid_search = GridSearchCV(pipeline, param_grid, cv=5)
grid_search.fit(X_train, y_train)

grid_search.best_params_
```

```python
final_model = LogisticRegression(penalty='elasticnet', solver='saga', C=100, l1_ratio=0.1, max_iter=500)
final_model.fit(X_train, y_train)
final_model.score(X_test, y_test)
```
